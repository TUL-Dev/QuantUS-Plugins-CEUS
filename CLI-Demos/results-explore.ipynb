```python
import os

curves_folder = '/home/das/Downloads/TTC-Results'
curves_dfs = []
for file in os.listdir(curves_folder):
    if file.endswith('.csv') and not os.path.basename(file).startswith('.'):
        df = pd.read_csv(os.path.join(curves_folder, file))
        curves_dfs.append(df)
```

```python
curves_dfs[0].head()
```

```python
import matplotlib.pyplot as plt

ix = 2
n_frames = 100
curve_name = 'moderate_diagnostics_Image-original_Mean'

for ix in range(len(curves_dfs)):
    scan_name = curves_dfs[ix]['Scan Name'].iloc[0]
    label = 'bolus' if 'bolus' in scan_name.lower() else 'molecular'
    time = curves_dfs[ix]['Time Array'].iloc[:n_frames]
    curve = curves_dfs[ix][curve_name].iloc[:n_frames]
    color = 'blue' if label == 'bolus' else 'orange'

    # plt.figure(figsize=(4, 3))
    # print(len(curve))
    plt.plot(time, curve, color=color, label=label)
plt.xlabel('Time (s)')
plt.xlim(0, 100)
# plt.ylabel(curve_name)
# plt.title(f'Curve for {scan_name} - {label}')
# plt.legend()
plt.show()
```

```python
vals = initial_df['Mean_moderate_diagnostics_Image-original_Mean']
labels = initial_df['contrast-type']

plt.figure(figsize=(8, 6))
for label in labels.unique():
    plt.hist(vals[labels == label], bins=15, alpha=0.5, label=label)
plt.xlabel('Mean_moderate_diagnostics_Image-original_Mean')
plt.ylabel('Frequency')
plt.title('Distribution by Contrast Type')
plt.legend()
plt.show()
```
